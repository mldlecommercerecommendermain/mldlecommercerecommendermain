```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt

# Algoritma adlarƒ± ve dosya e≈üle≈ümeleri
algoritma_dosyalar = {
    "Random": "random_based_sonuclari.pkl",
    "Content-Based": "contant_based_sonuclari.pkl",
    "Item-Base": "item_based_sonuclari.pkl",
    "Hybrid": "hybrid_sonuclari.pkl",
    "RNN": "rnn_sonuclari.pkl",
    "Random Forest": "random_forest_sonuclari.pkl",
     "Popularity": "popularity_sonuclari.pkl"
}

# Sonu√ßlarƒ± depolamak i√ßin bo≈ü dict
sonuclar = {}

# T√ºm dosyalarƒ± y√ºkle
for isim, dosya in algoritma_dosyalar.items():
    with open(dosya, "rb") as f:
        metrikler = pickle.load(f)
        sonuclar[isim] = metrikler
        

# Pandas DataFrame'e √ßevir
df_sonuclar = pd.DataFrame(sonuclar).T.round(4) # Transpose: algoritmalar satƒ±ra ge√ßsin

# Tabloda g√∂ster
print("üîç Algoritma Performans Sonu√ßlarƒ±:")
display(df_sonuclar)
```

```python
import matplotlib.pyplot as plt

# MAP
plt.figure(figsize=(8, 5))
df_sonuclar["MAP"].plot(kind="bar", title="MAP Kar≈üƒ±la≈ütƒ±rmasƒ±", ylabel="MAP")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```
